```python
# load the dataset into DataFram
df=pd.read_csv('Indiana_population_citizen.csv')
```

```python
# set the display option
pd.set_option('display.max_rows',20)
pd.set_option('display.precision',4)
df.head()
```

```python
# select top 10 populated States
df_top_3=df[0:3]
df_top_3
```

```python
# Original Barchart
axes=df_top_3.plot.bar()
axes.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.xticks(np.arange(3), ('California', 'Texas', 'Florida'), rotation=0)
plt.title('Immigration Status by Top 3 States in the United States')
plt.show()
```

```python
# Revised Barchart
sns.set_palette('RdBu')
sns.set_style('darkgrid')
fig,(ax0,ax1,ax2,ax3,ax4) =plt.subplots(nrows=1, ncols=5, sharey=True, figsize=(20,5))
sns.barplot(y='Total Population', x='State', data=df_top_3, ax=ax0)
ax0.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax0.set_title('Total Population - Top 3 States')
sns.barplot(y='US Citizen Born in the United States', x='State', data=df_top_3, ax=ax1)
ax1.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax1.set_title('US Citizen Born in the United States- Top 3 States')
sns.barplot(y='US Citizen by Naturalization', x='State', data=df_top_3, ax=ax2)
ax2.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax2.set_title('US Citizen by Naturalization - Top 3 States')
sns.barplot(y='US Citizen born abroad', x='State', data=df_top_3, ax=ax3)
ax3.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax3.set_title('US Citizen born abroad - Top 3 States')
sns.barplot(y='Not a US Citizen', x='State', data=df_top_3, ax=ax4)
ax4.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax4.set_title('Not a US Citizen - Top 3 States')
plt.show()
```
