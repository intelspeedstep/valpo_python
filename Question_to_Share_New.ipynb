```python
# load the dataset into DataFram
df=pd.read_csv('Indiana_population_citizen.csv')
```

```python
# set the display option
pd.set_option('display.max_rows',20)
pd.set_option('display.precision',4)
df.head()
```

```python
# select top 10 populated States
df_top_3=df[0:3]
df_top_3=df_top_3.drop('US Citizen born abroad',axis=1)
```

```python
# Original Barchart
axes=df_top_3.plot.bar()
axes.get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.xticks(np.arange(3), ('California', 'Texas', 'Florida'), rotation=0)
plt.title('Immigration Status by Top 3 States in the United States')
plt.show()
```

```python
# Revised Barchart
sns.set_palette('RdBu')
sns.set_style('darkgrid')
fig,axes =plt.subplots(nrows=2, ncols=2, sharey=True,figsize=(10, 10))
sns.barplot(y='Total Population', x='State', data=df_top_3, ax=axes[0,0])
axes[0,0].get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
axes[0,0].set_title('Total Population - Top 3 States')
sns.barplot(y='US Citizen Born in the United States', x='State', data=df_top_3, ax=axes[0,1])
axes[0,1].get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
axes[0,1].set_title('US Citizen Born in the United States- Top 3 States')
sns.barplot(y='US Citizen by Naturalization', x='State', data=df_top_3, ax=axes[1,0])
axes[1,0].get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
axes[1,0].set_title('US Citizen by Naturalization - Top 3 States')
sns.barplot(y='Not a US Citizen', x='State', data=df_top_3, ax=axes[1,1])
axes[1,1].get_yaxis().set_major_formatter(matplotlib.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
axes[1,1].set_title('Not a US Citizen - Top 3 States')
plt.show()
```
